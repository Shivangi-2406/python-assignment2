```python
# library management system
def add_book(title, author, genre):
    return (title, author, genre)


def add_to_library(book, library, book_set):
    if book not in book_set:
        library.append(book)
        book_set.add(book)
        print(f"Book '{book[0]}' by {book[1]} added to the library.")
    else:
        print(f"Book '{book[0]}' by {book[1]} already exists in the library.")


def remove_from_library(title, library, book_set):
    for book in library:
        if book[0] == title:
            library.remove(book)
            book_set.remove(book)
            print(f"Book '{title}' removed from the library.")
            return
    print(f"Book '{title}' not found in the library.")


def search_books(search_term, library):
    results = [book for book in library if search_term.lower() in book[0].lower() or search_term.lower() in book[1].lower()]
    return results


def list_books(library):
    if library:
        print("Books in the library:")
        for book in library:
            print(f"Title: {book[0]}, Author: {book[1]}, Genre: {book[2]}")
    else:
        print("No books in the library.")


def categorize_books(library):
    genre_dict = {}
    for book in library:
        genre = book[2]
        if genre not in genre_dict:
            genre_dict[genre] = []
        genre_dict[genre].append(book)
    return genre_dict


library = []
book_set = set()


book1 = add_book("The martian", "by Andy Weir", " science Fiction")
book2 = add_book("The book thief", "Markus Zusak", " historical Fiction")
book3 = add_book("Bird Box", "Josh Malerman", "Horror")
book4 = add_book("The Great Gatsby", "F. Scott Fitzgerald", "Fiction")  

add_to_library(book1, library, book_set)
add_to_library(book2, library, book_set)
add_to_library(book3, library, book_set)
add_to_library(book4, library, book_set)  


remove_from_library("The book thief", library, book_set)


search_results = search_books("Great", library)
print("Search results:")
for book in search_results:
    print(f"Title: {book[0]}, Author: {book[1]}, Genre: {book[2]}")


list_books(library)


genre_categories = categorize_books(library)
print("Books categorized by genre:")
for genre, books in genre_categories.items():
    print(f"Genre: {genre}")
    for book in books:
        print(f"  Title: {book[0]}, Author: {book[1]}")
```
